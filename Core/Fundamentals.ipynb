### Reshaping, Stacking, Squeezing, and Unsqueezing

```python
# Create a tensor
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped
```

```python
# Add an extra dimension
x_reshaped = x.reshape(9, 1)
x_reshaped
```

```python
# Change the view (shares the same memory as x)
z = x.view(1, 9)
z, z.shape
```

```python
# changing z changes x
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```

```python
# Squeeze tensor
x_reshaped.shape, x_reshaped.squeeze()
```

```python
x_reshaped.squeeze().shape
```

```python
x_squeezed = x_reshaped.squeeze()
x_squeezed
```

```python
# Unsqueeze - add a single dimension
x_unsqueezed = x_squeezed.unsqueeze(dim=1)
x_unsqueezed
```

```python
# Permute - rearrange the dimensions
x_original = torch.rand(size=(224, 224, 3))
print(x_original.shape)

# Permute the original tensor
x_permuted = x_original.permute(2, 0, 1)
x_permuted.shape
```

### Pytorch Tensors and NumPy

```python
# NumPy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)  # warning: when converting from nupy to pytorch, pytorch reflects numpy's default (float64)
array, tensor
```

```python
array.dtype, tensor.dtype
```

```python
array = array + 1
```

```python
array, tensor
```

```python
# Tensor to numpy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
tensor.dtype, numpy_tensor.dtype
```

```python
tensor = tensor + 1
tensor, numpy_tensor
```

### Reproducibility (trying to take random out of random)

```python
rand_a = torch.rand(3, 4)
rand_b = torch.rand(3, 4)
rand_a == rand_b
```

```python
# random but reproducible tensors
# set seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
rand_c = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
rand_d = torch.rand(3, 4)
rand_c == rand_d
```

```python
# Check GPU access
import torch
torch.cuda.is_available()
```

```python
# Setup device agnostic code
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
# Count number of devices
torch.cuda.device_count()
```

```python
# Putting tensors (and models) on the GPU
tensor = torch.tensor([1, 2, 3])

# Tensor not on GPU
print(tensor, tensor.device)
```

```python
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```

```python
# Putting tensors (and models) on the GPU
tensor = torch.tensor([1, 2, 3], device=device)

# Tensor not on GPU
print(tensor, tensor.device)
```

```python
# Move back to the CPU
tensor_back_on_cpu = tensor_on_gpu.cpu().numpy()
tensor_back_on_cpu
```
